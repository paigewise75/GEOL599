```python
import numpy as np
from sklearn.decomposition import PCA
import pandas as pd
import matplotlib.pyplot as plt


filename = "Waves_2023.txt"
file = pd.DataFrame(pd.read_csv(filename, delim_whitespace=True))

# Drop 999, 99, and NA values
df = file[['WVHT', 'APD', 'DPD', 'MWD', 'WTMP']].replace([9999.0,999.0, 99.0, 'NA'], np.nan).dropna()
df.shape
```

```python
pca = PCA(n_components=5)
pca.fit(df)

print(pca.explained_variance_ratio_)

```

```python
print(np.abs(pca.components_))
plotData = pca.transform(df)
```

```python
plt.scatter(plotData[:, 0], plotData[:, 1], color='purple', edgecolors='white')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Bouy Data')
plt.show()
```

```python
print(" 2) Number of principal components needed to capture 99% of the variability in the data is 1")

print(" 3) The first principal component explains 99.9% of the variability in the data and the second principal component explains 0.1% of the variability in the data. The first principal component is the mean wave direction and the second principal component is the wave height.")

print(" 4) It tells you that the majority of the variability in the data is captured by the first principal component which is mean wave direction.")
```
